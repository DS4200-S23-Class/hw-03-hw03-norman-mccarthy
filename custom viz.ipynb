```python
import pandas as pd
```

```python
data = pd.read_csv('rotten_tomatoes_top_movies.csv')
```

```python
data = data[['critic_score', 'people_score', 'type']]
```

```python
data
```

```python
import matplotlib.pyplot as plt
```

```python
x = data['critic_score']
y = data['people_score']
groups = data.groupby('type')
for name, group in groups:
    plt.rcParams['figure.facecolor'] = 'white'
    plt.plot(group.critic_score, group.people_score, marker='o', linestyle='', markersize=5, label=name)
ax = plt.subplot(111)
# Shrink current axis by 20%
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
# Put a legend to the right of the current axis
ax.legend(title='Genre type', loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('Critic Scores vs People Scores')
plt.xlabel('Critic Score')
plt.ylabel('People score')
ax.set_facecolor('xkcd:white')
plt.show()
```
